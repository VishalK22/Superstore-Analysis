```python
# Seprating the year from Ship Date column
cr['year'] = pd.DatetimeIndex(cr['Ship Date']).year
cr.head()
```

```python
# Unique value count in every year
cr["year"].value_counts()
```

```python
# Analyzing the Profit and Discount column in every year 
cr.groupby('year')['Profit','Discount'].agg(['sum']).plot.bar()
plt.title('Total Profit and Discount per Sub-Category')
plt.show()
```

```python
cr.groupby('year')['Profit'].agg(['sum']).plot.bar()
plt.title('Total Profit per Year')
plt.show()
```

```python
cr.groupby('year')['Discount'].agg(['sum']).plot.bar()
plt.title('Total Discount per Year')
plt.show()
```

```python
cr.groupby('year')['Profit','Discount'].agg(['sum']).plot.barh(width=1, figsize=(18,5))
plt.title('Total Profit and Discount per Sub-Category')
plt.show()
```
